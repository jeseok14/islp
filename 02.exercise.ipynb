### 3. We now revisit the bias-variance decomposition.

#### a) Provide a sketch of typical (squared) bias, variance, training error, test error, and Bayes (or irreducible) error curves, on a single plot, as we go from less flexible statistical learning methods towards more flexible approaches. The x-axis should represent the amount of flexibility in the method, and the y-axis should represent the values for each curve. There should be five curves. Make sure to label each one.

```python
import numpy as np
import matplotlib.pyplot as plt
import koreanize_matplotlib

x = np.arange(0.5, 10, 0.01)

def bias(data):
    return 1/data

plt.xlabel('복잡도')
plt.plot(x, bias(x), label='squared bias')
plt.plot(x, bias(x)[::-1], label='variance')
plt.plot(x, bias(x) + bias(x)[::-1], label = 'training error')
plt.plot(x, bias(x) + bias(x)[::-1] + np.ones_like(x), label = 'test error')
plt.plot(x, np.ones_like(x), label='irriducible error')

plt.title('대략적인 모델의 복잡도에 따른 양상')
plt.legend()
plt.show()
```

#### b) Explain why each of the five curves has the shape displayed in part (a).

- bias : 모델이 복잡해질수록 , 더 유연해질수록 bias는 줄어든다. - 유연성이 증가함에 따라 감소한다.
- variance : 모델이 복잡해질수록, 더 유연해질수록 variance는 늘어난다. - 모델의 유연성에 대해 단조 증가한다.
- training error : 훈련 데이터에 대한 MSE $MSE = \frac 1n \sum_{i = 1}^{n}(y_i - \hat f(x_i))^2$로 모델이 유연해짐에 따라 감소한다.
- test error : 훈련 관측값이 아닌 테스트 데이터 $(x_0, y_0)$에 대한 제곱예측오차 $E(y_0 - \hat f(x_0))^2 = Var(\hat f(x_0)) + [Bias(\hat f(x_0))]^2 + Var(\epsilon)$이고, 이 값들은 모두 음수가 아니다. 여기서 유연성이 증가하면 오류가 감소하다가, 최소값을 지난 이후로 다시 증가하게 된다. 이 최소값은 irreducible error보다 항상 크거나 같게 된다.
